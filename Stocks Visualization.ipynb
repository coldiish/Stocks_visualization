# Finance Data Project

```python
# Required imports
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
```

**The data used in this project is of American banks from the year 2006 to 2016.
Source : Yahoo Finance**

```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
# Bank of America
BAC = data.DataReader('BAC', 'yahoo', start, end)

# CitiGroup
C = data.DataReader('C', 'yahoo', start, end)

# Goldman Sachs
GS = data.DataReader('GS', 'yahoo', start, end)

# JPMorgan Chase
JPM = data.DataReader('JPM', 'yahoo', start, end)

# Morgan Stanley
MS = data.DataReader('MS', 'yahoo', start, end)

# Wells Fargo
WFC = data.DataReader('WFC', 'yahoo', start, end)
```

```python
BAC.head()
```

**List of tickers for the banks.**

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

**Combining the individual data of the banks into a single dataframe.**

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys=tickers)
```

```python
bank_stocks
```

**Setting column names.**

```python
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
```

```python
bank_stocks.head()
```

## Maximum closing price for each bank

```python
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

**Creating a new dataframe 'returns' that stores the percent change in price.**

```python
returns = pd.DataFrame()
```

```python
for tick in tickers: 
    returns[tick+' Returns'] = bank_stocks[tick]['Close'].pct_change()
```
